````python
import pandas as pd
import requests
pd.set_option("display.min_rows", 500)

# 1) seu código aqui
sinasc = pd.read_csv('SINASC_RO_2019.csv')
print('Número de registros................: ',sinasc.shape)
print('Número de registros não duplicados.: ',sinasc.drop_duplicates().shape)
# Não há duplicados
````

````python
# 2) seu código aqui
sinasc.isna().sum()
````

````python
# 3) seu código aqui
# Criando a Base
dfsinasc = sinasc[['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 
    'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']]
# Verificando Missing
dfsinasc.isna().sum() 
````

````python
# 4) seu código aqui
dfsinasc = dfsinasc.dropna(subset=['APGAR5'])
dfsinasc.isna().sum() 
````

````python
# 5) seu código aqui
dfsinasc
dfsinasc['ESTCIVMAE'] = dfsinasc['ESTCIVMAE'].fillna(9)
dfsinasc['CONSULTAS'] = dfsinasc['CONSULTAS'].fillna(9)
````

````python
# 6) Seu código aqui
dfsinasc['QTDFILVIVO'] = dfsinasc['QTDFILVIVO'].fillna(0)
dfsinasc
````

````python
# 7) seu código aqui
#Criando um dfMae para testes 
dfMae = sinasc[['IDADEMAE', 'ESTCIVMAE', 'ESCMAE']]

# Verificando quais os Estados civis mais utilizados
dfMae[['IDADEMAE', 'ESTCIVMAE', 'ESCMAE']].value_counts()
dfMae.sort_values(by=['IDADEMAE'])

# De acordo com a lei, somente maiores de 16 anos com autorização podem se casar, sendo assim
# idades < 16 é mais coerente colocar Estado civil 1 - Solteiro
# Aplicando o FillNa = 1 - Solteiro na base sinasc
dfsinasc[(dfsinasc['IDADEMAE']<16) & (dfsinasc['ESTCIVMAE'].isna())].fillna(1)
dfsinasc
````

````python
# idades >= 16 acho mais coerente colocar o que é verdadeiro, ou seja, 9-ignorado, para não ter relevância com dados falsos
# dfMae16 para testes
dfMae16 = sinasc[['IDADEMAE', 'ESTCIVMAE', 'ESCMAE']]

#Aplicando o FillNa = 9 - Ignorado na base sinasc
dfsinasc[(dfsinasc['IDADEMAE']>=16) & (dfsinasc['ESTCIVMAE'].isna())].fillna(9)
````

````python
# Verificando os campos isNa Restantes
dfsinasc.isna().sum() 
````

````python
# [Campo : Escolaridade Mãe]
# O mais coerente colocar o que é verdadeiro, ou seja, 9-ignorado, para não ter relevância com dados falsos
#Aplicando o FillNa = 9 - Ignorado na base sinasc
dfsinasc['ESCMAE'] = dfsinasc['ESCMAE'].fillna(9)
dfsinasc
````

````python
# Verificando os campos isNa Restantes
dfsinasc.isna().sum() 
````

````python
# [Campo : GESTACAO]
# Verificando Qual o tipo de Gestação mais comum em %
dfsinasc.value_counts('GESTACAO') / dfsinasc.shape[0] *100
````

````python
# Aplicando as semanas mais comuns de gestação no isNa
dfsinasc['GESTACAO'] = dfsinasc['GESTACAO'].fillna('37 a 41 semanas')
dfsinasc
````

````python
# Verificando os campos isNa Restantes
dfsinasc.isna().sum() 
````

````python
# [Campo : GRAVIDEZ]
# Verificando Qual o tipo de GRAVIDEZ mais comum em %
dfsinasc.value_counts('GRAVIDEZ') / dfsinasc.shape[0] *100
````

````python
# Aplicando o tipo de GRAVIDEZ mais comun no isNa
dfsinasc['GRAVIDEZ'] = dfsinasc['GRAVIDEZ'].fillna('Única')
dfsinasc
````

````python
# Verificando os campos isNa Restantes
dfsinasc.isna().sum() 
#Não existem mais campos IsNa
````

````python
# 8) seu código aqui
# Entre 8 e 10 está em uma faixa 'normal'.
# Entre 6 e 7, significa que o recém-nascido passou por 'asfixia leve'.
# Entre 4 e 5 significa 'asfixia moderada'.
# Entre 0 e 3 significa 'asfixia severa'.

dfsinasc.loc[(dfsinasc['APGAR5'] >=8) & (dfsinasc['APGAR5'] <=10), 'Categoria'] = 'Normal'
dfsinasc.loc[(dfsinasc['APGAR5'] >=6) & (dfsinasc['APGAR5'] <=7), 'Categoria']  = 'asfixia leve'
dfsinasc.loc[(dfsinasc['APGAR5'] >=4) & (dfsinasc['APGAR5'] <=5), 'Categoria'] = 'asfixia moderada'
dfsinasc.loc[(dfsinasc['APGAR5'] >=0) & (dfsinasc['APGAR5'] <=3), 'Categoria'] = 'asfixia severa'

dfsinasc
````

````python
#Frequência
frequencias = dfsinasc.groupby('Categoria').size().reset_index(name='frequencia')

# Exibindo as frequências
frequencias
````
