```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests

%matplotlib inline
# %matplotlib notebook

sinasc = pd.read_csv('sinasc.csv')
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
plt.figure(figsize=(8, 6))  # Definir o tamanho da figura (opcional)
sns.set(style="whitegrid")  # Estilo do grid (opcional)
sns.boxplot(x='IDADEMAE', data=sinasc)
plt.title('Boxplot de Idade')
plt.xlabel('Idade')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
sns.boxplot(x='PESO', data=sinasc)
plt.title('Boxplot do Peso do Bebê')
plt.xlabel('Peso')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
sinasc['IDADEMAE'].plot.hist(bins = 20)
plt.title('Histplot de Idade da Mãe')
plt.xlabel('Idade da mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")
sinasc['PESO'].plot.hist(bins = 20)
plt.title('Histplot de Peso do bebê')
plt.xlabel('Peso do bebÊ')
plt.ylabel('Frequência')
plt.show()
```
