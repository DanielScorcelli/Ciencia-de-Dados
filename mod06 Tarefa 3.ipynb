```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests

%matplotlib inline
# %matplotlib notebook

sinasc = pd.read_csv('SINASC_RO_2019.csv')
```

#### 2) Categorize a variável APGAR5 conforme as seguintes quebras:

- 0 a 3: 'asfixia grave'
- 4 a 5: 'asfixia moderada'
- 6 a 7: 'asfixia leve'
- 8 a 10: 'normal'

```python
intervalos = [0, 3, 5, 7, 10]
rotulos = ['Asfixia Grave', 'Asfixia Moderada', 'Asfixia Leve', 'Normal']
sinasc['APGAR5_categorias'] = pd.cut(sinasc['APGAR5'], bins=intervalos, labels=rotulos, include_lowest=True)
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.countplot(x='APGAR5_categorias', data=sinasc, palette='viridis')
plt.title('Distribuição da variável APGAR5 categorizada')
plt.xlabel('Categorias de APGAR5')
plt.ylabel('Contagem')

plt.show()

```

#### 3) Monte um gráfico comparando a distribuição de APGAR5 por sexo para cada grupo de tempo de gestação (GESTACAO).

```python
#Ordenando o eiso X
ORDEM = ['Menos de 22 semanas','22 a 27 semanas','28 a 31 semanas', '32 a 36 semanas', '37 a 41 semanas', '42 semanas e mais'] 

plt.figure(figsize=(12, 8))
sns.barplot(x='GESTACAO', y='APGAR5', hue='SEXO', data=sinasc, errorbar=None, order=ORDEM)

plt.title('Distribuição de APGAR5 por Sexo e Tempo de Gestação')
plt.xlabel('Tempo de Gestação')
plt.ylabel('APGAR5')
plt.legend(title='Sexo')



plt.show()
```

####  4) Monte uma tabela cruzada com APGAR5 nas colunas e GESTACAO nas linhas

```python
# Tabela cruzada
freq = pd.crosstab(sinasc['GESTACAO'], sinasc['APGAR5'])
freq = freq.sort_index()
freq
```

####  5) Da tabela cruzada do intem anterior, calcule os percentuais de linha. Ou seja, a distribuição da variável APGAR5 para cada classe de GESTACAO dado.

```python
freq_pctlin = freq.div(freq.sum(axis = 1), axis = 0)
freq_pctlin = freq_pctlin.sort_index()

freq_pctlin
```
