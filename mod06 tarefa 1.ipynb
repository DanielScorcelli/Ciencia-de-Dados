#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests

%matplotlib inline
# %matplotlib notebook

sinasc = pd.read_csv('sinasc.csv')
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python

plt.rc('figure', figsize=(10, 8))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=sinasc)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=sinasc)
sns.countplot(ax = axes[1, 0], x='SEXO', data=sinasc)
sns.countplot(ax = axes[1, 1], x='GESTACAO', data=sinasc)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=sinasc)
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python

plt.rc('figure', figsize=(6, 4))
sns.barplot(x='APGAR5', y='CONSULTAS', data=sinasc)
plt.xlabel('APGAR5')
plt.ylabel('Número de Consultas')
plt.title('Relação entre APGAR5 e Número de Consultas')

plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.rc('figure', figsize=(12, 8))
fig, axes = plt.subplots(2, 2)
fig.tight_layout(pad=12.0)



sns.countplot(ax = axes[0, 0], x='SEXO', data=sinasc)
axes[0][0].tick_params(axis='x', rotation=90)

sns.countplot(ax = axes[0, 1], x='GESTACAO', data=sinasc)
axes[0][1].tick_params(axis='x', rotation=90)

sns.countplot(ax = axes[1, 0], x='ESTCIVMAE', data=sinasc)

sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=sinasc)

```
